### 14. Remove Duplicates from a List

```python
def remove_duplicates(input_list):
    # Using a set to efficiently store unique elements
    return list(set(input_list))

my_list_with_duplicates = [1, 2, 2, 3, 4, 4, 5, 1, 6]
print(f"Original list: {my_list_with_duplicates}")
print(f"List with duplicates removed: {remove_duplicates(my_list_with_duplicates)}")

string_list_duplicates = ['apple', 'banana', 'apple', 'orange', 'banana']
print(f"Original string list: {string_list_duplicates}")
print(f"String list with duplicates removed: {remove_duplicates(string_list_duplicates)}")
```

### 15. Find Common Elements in Two Lists

```python
def find_common_elements(list1, list2):
    # Convert lists to sets for efficient intersection operation
    return list(set(list1) & set(list2))

list_a = [1, 2, 3, 4, 5]
list_b = [4, 5, 6, 7, 8]
print(f"List A: {list_a}")
print(f"List B: {list_b}")
print(f"Common elements: {find_common_elements(list_a, list_b)}")

list_c = ['red', 'green', 'blue']
list_d = ['blue', 'yellow', 'green']
print(f"List C: {list_c}")
print(f"List D: {list_d}")
print(f"Common elements: {find_common_elements(list_c, list_d)}")
```

### 16. Merge Two Dictionaries

```python
def merge_dictionaries(dict1, dict2):
    # Using the update method (modifies dict1 in place)
    # merged_dict = dict1.copy()
    # merged_dict.update(dict2)
    # return merged_dict

    # Using the ** operator (Python 3.5+), creates a new dictionary
    return {**dict1, **dict2}

dict_a = {'a': 1, 'b': 2}
dict_b = {'c': 3, 'd': 4}
dict_c = {'a': 10, 'e': 5} # 'a' will be overwritten by dict_c's value

print(f"Dictionary A: {dict_a}")
print(f"Dictionary B: {dict_b}")
print(f"Merged A and B: {merge_dictionaries(dict_a, dict_b)}")

print(f"Dictionary C: {dict_c}")
print(f"Merged A and C (with overwrite): {merge_dictionaries(dict_a, dict_c)}")
```

### 17. Count Word Frequencies in a String

```python
from collections import Counter
import re # For splitting by non-alphanumeric characters

def count_word_frequencies(text):
    # Convert to lowercase and split by non-word characters
    words = re.findall(r'\b\w+\b', text.lower())
    return Counter(words)

sentence = "This is a test. This test is only a test."
print(f"Original sentence: '{sentence}'")
print(f"Word frequencies: {count_word_frequencies(sentence)}")

example_text = "Apple Banana Apple orange BANANA"
print(f"Original text: '{example_text}'")
print(f"Word frequencies: {count_word_frequencies(example_text)}")
```

### 18. Check for Anagrams

```python
def are_anagrams(str1, str2):
    # Normalize strings (remove spaces, convert to lowercase)
    str1_normalized = sorted(str1.replace(" ", "").lower())
    str2_normalized = sorted(str2.replace(" ", "").lower())

    return str1_normalized == str2_normalized

print(f"'listen' and 'silent' are anagrams: {are_anagrams('listen', 'silent')}")
print(f"'Debit Card' and 'Bad Credit' are anagrams: {are_anagrams('Debit Card', 'Bad Credit')}")
print(f"'hello' and 'world' are anagrams: {are_anagrams('hello', 'world')}")
```

### 19. Sort a List of Dictionaries by a Key

### 24. Class Variables vs. Instance Variables

```python
class Dog:
    # Class variable: shared by all instances of the class
    species = "Canis familiaris"

    def __init__(self, name, age):
        # Instance variables: unique to each instance
        self.name = name
        self.age = age

    def description(self):
        return f"{self.name} is {self.age} years old."

# Creating instances
dog1 = Dog("Buddy", 3)
dog2 = Dog("Lucy", 5)

print(f"Dog 1: {dog1.description()}, Species: {dog1.species}")
print(f"Dog 2: {dog2.description()}, Species: {dog2.species}")

# Accessing class variable via class name
print(f"Class species: {Dog.species}")

# Modifying class variable (affects all instances and the class itself)
Dog.species = "Domestic Dog"
print(f"Dog 1 species after modification: {dog1.species}")
print(f"Dog 2 species after modification: {dog2.species}")

# If you assign to dog1.species, it creates an instance variable named 'species' for dog1,
# shadowing the class variable for that instance, but not affecting the class or other instances.
dog1.species = "Labrador"
print(f"Dog 1's species (instance variable): {dog1.species}")
print(f"Dog 2's species (still class variable): {dog2.species}")
print(f"Class species (still changed class variable): {Dog.species}")
```

### 25. `__str__` and `__repr__` (Object Representation)

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        # User-friendly string representation (for end-users)
        return f"({self.x}, {self.y})"

    def __repr__(self):
        # Developer-friendly string representation (unambiguous, recreate object)
        return f"Point(x={self.x}, y={self.y})"

p = Point(10, 20)

print(f"Using str(): {str(p)}") # Called by print(), str()
print(f"Using repr(): {repr(p)}") # Called by repr(), debugger

# When called directly, __str__ is preferred if available
print(p)

# In lists/collections, __repr__ is often used
point_list = [Point(1,1), Point(2,2)]
print(f"List of points: {point_list}")
```

### 26. Static Methods and Class Methods

```python
class MathOperations:
    _pi = 3.14159 # 'Protected' class variable

    def __init__(self, value):
        self.value = value

    @staticmethod
    def add_numbers(a, b):
        # Static methods don't need access to instance (self) or class (cls)
        # They are like regular functions but logically belong to the class.
        return a + b

    @classmethod
    def get_pi(cls):
        # Class methods receive the class itself (cls) as the first argument.
        # They can access/modify class variables or create instances.
        return cls._pi

    @classmethod
    def create_zero_instance(cls):
        # Factory method using a class method
        return cls(0)

# Accessing static method via class name (or instance, though less common)
print(f"Static method add_numbers(5, 3): {MathOperations.add_numbers(5, 3)}")

# Accessing class method via class name
print(f"Class method get_pi(): {MathOperations.get_pi()}")

# Creating an instance using a class method
zero_op = MathOperations.create_zero_instance()
print(f"Instance created by class method: value = {zero_op.value}")

# Instance can also call class/static methods
my_op = MathOperations(10)
print(f"Instance calling static method add_numbers(2, 7): {my_op.add_numbers(2, 7)}")
print(f"Instance calling class method get_pi(): {my_op.get_pi()}")
```

### 27. Properties (Getters and Setters)

```python
class Temperature:
    def __init__(self, celsius):
        self._celsius = None # Initialize internal storage
        self.celsius = celsius # Use the setter to validate

    @property
    def celsius(self):
        # The getter method
        return self._celsius

    @celsius.setter
    def celsius(self, value):
        # The setter method with validation
        if not isinstance(value, (int, float)):
            raise TypeError("Celsius temperature must be a number.")
        if value < -273.15:
            raise ValueError("Temperature cannot be below absolute zero (-273.15 C).")
        self._celsius = value

    @property
    def fahrenheit(self):
        # A read-only property derived from celsius
        return (self.celsius * 9/5) + 32

    @fahrenheit.setter
    def fahrenheit(self, value):
        # Setter for fahrenheit (converts to celsius internally)
        if not isinstance(value, (int, float)):
            raise TypeError("Fahrenheit temperature must be a number.")
        self.celsius = (value - 32) * 5/9 # Use the celsius setter for validation

# Creating an instance
temp = Temperature(25)

print(f"Initial Celsius: {temp.celsius}°C")
print(f"Initial Fahrenheit: {temp.fahrenheit}°F")

# Setting temperature via property
temp.celsius = 30
print(f"Updated Celsius: {temp.celsius}°C")
print(f"Updated Fahrenheit: {temp.fahrenheit}°F")

# Setting temperature via Fahrenheit property
temp.fahrenheit = 212 # Boiling point
print(f"Updated Celsius: {temp.celsius}°C")
print(f"Updated Fahrenheit: {temp.fahrenheit}°F")

# Demonstrating validation
try:
    temp.celsius = -300 # Below absolute zero
except ValueError as e:
    print(f"Error setting temperature: {e}")

try:
    temp.celsius = "hot"
except TypeError as e:
    print(f"Error setting temperature: {e}")
```

### 2. Array Creation

NumPy's core is the `ndarray` object. Here's how to create them in various ways:

```python
# From a Python list
arr1 = np.array([1, 2, 3, 4, 5])
print(f"Array from list: {arr1}\n")

# From nested Python lists (2D array)
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print(f"2D Array from nested lists:\n{arr2}\n")

# Array of zeros
zeros_arr = np.zeros((2, 3)) # 2 rows, 3 columns
print(f"Array of zeros:\n{zeros_arr}\n")

# Array of ones
ones_arr = np.ones((3, 2))
print(f"Array of ones:\n{ones_arr}\n")

# Empty array (content is random, depends on memory)
empty_arr = np.empty((2, 2))
print(f"Empty array (random content):\n{empty_arr}\n")

# Array with a range of numbers
range_arr = np.arange(0, 10, 2) # start, stop (exclusive), step
print(f"Array with range (step 2): {range_arr}\n")

# Array with evenly spaced numbers
linspace_arr = np.linspace(0, 10, 5) # start, stop (inclusive), number of elements
print(f"Array with linspace (5 elements): {linspace_arr}\n")

# Identity matrix
identity_matrix = np.eye(3) # 3x3 identity matrix
print(f"Identity matrix (3x3):\n{identity_matrix}\n")

# Random arrays
rand_uniform = np.random.rand(2, 2) # Uniform distribution [0, 1)
print(f"Random uniform 2x2 array:\n{rand_uniform}\n")

rand_normal = np.random.randn(2, 2) # Standard normal distribution (mean=0, var=1)
print(f"Random normal 2x2 array:\n{rand_normal}\n")
```

### 7. Grouping and Aggregation

Using `groupby()` to aggregate data by categories.

```python
products_data = {
    'Category': ['Electronics', 'Clothes', 'Electronics', 'Clothes', 'Electronics', 'Books'],
    'Item': ['Laptop', 'Shirt', 'Mouse', 'Pants', 'Keyboard', 'Novel'],
    'Price': [1200, 50, 25, 75, 80, 15],
    'Quantity': [2, 10, 5, 8, 3, 20]
}
df_products = pd.DataFrame(products_data)
print("Original Products DataFrame:\n")
display(df_products)

# Group by 'Category' and calculate the sum of 'Price' and 'Quantity'
print("\nSum of Price and Quantity by Category:\n")
display(df_products.groupby('Category').sum(numeric_only=True))

# Group by 'Category' and calculate the mean of 'Price'
print("\nMean Price by Category:\n")
display(df_products.groupby('Category')['Price'].mean())

# Group by 'Category' and apply multiple aggregation functions
print("\nMultiple aggregations by Category:\n")
display(df_products.groupby('Category').agg(
    Total_Price=('Price', 'sum'),
    Avg_Quantity=('Quantity', 'mean'),
    Count_Items=('Item', 'count')
))
```

### 8. Merging DataFrames

Combining DataFrames based on common columns (like SQL joins).

```python
df_orders = pd.DataFrame({
    'OrderID': [1, 2, 3, 4],
    'CustomerID': [101, 102, 101, 103],
    'Product': ['A', 'B', 'C', 'A']
})
df_customers = pd.DataFrame({
    'CustomerID': [101, 102, 103, 104],
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Email': ['a@example.com', 'b@example.com', 'c@example.com', 'd@example.com']
})

print("Orders DataFrame:\n")
display(df_orders)
print("\nCustomers DataFrame:\n")
display(df_customers)

# Inner merge (default) - only common CustomerIDs
df_inner_merge = pd.merge(df_orders, df_customers, on='CustomerID', how='inner')
print("\nInner Merge (orders with matching customer info):\n")
display(df_inner_merge)

# Left merge - keep all orders, add customer info if available
df_left_merge = pd.merge(df_orders, df_customers, on='CustomerID', how='left')
print("\nLeft Merge (all orders, customer info if exists):\n")
display(df_left_merge)

# Right merge - keep all customers, add order info if available
df_right_merge = pd.merge(df_orders, df_customers, on='CustomerID', how='right')
print("\nRight Merge (all customers, order info if exists):\n")
display(df_right_merge)

# Outer merge - keep all records from both, fill NaN where no match
df_outer_merge = pd.merge(df_orders, df_customers, on='CustomerID', how='outer')
print("\nOuter Merge (all records from both dataframes):\n")
display(df_outer_merge)
```

### Introduction to SQL

SQL (Structured Query Language) is a standard language for storing, manipulating and retrieving data in databases. Here are some basic, yet fundamental, queries.

### 1. Data Definition Language (DDL)

DDL commands are used to define database structures like tables.

#### `CREATE TABLE`

Creates a new table in the database. You specify the table name and then define each column's name, data type, and any constraints (e.g., `PRIMARY KEY`, `NOT NULL`).

```python
%%sql
CREATE TABLE Employees (
    EmployeeID INT PRIMARY KEY,
    FirstName VARCHAR(50) NOT NULL,
    LastName VARCHAR(50) NOT NULL,
    Department VARCHAR(50),
    HireDate DATE,
    Salary DECIMAL(10, 2)
);
```

#### `ALTER TABLE` and `DROP TABLE`

- `ALTER TABLE` is used to add, delete, or modify columns in an existing table.
- `DROP TABLE` deletes an existing table, including all its data and structure.

```python
%%sql
-- Add a new column
ALTER TABLE Employees
ADD Email VARCHAR(100);

-- Drop a table
-- DROP TABLE Employees;
```

### 2. Data Manipulation Language (DML)

DML commands are used for managing data within schema objects. These are the most common queries for interacting with data.

#### `INSERT INTO`

Adds new rows of data into a table.

#### `SELECT`

Retrieves data from a database. This is the most frequently used SQL command.

##### Select all columns (`*`)

```python
%%sql
SELECT * FROM Employees;
```

##### Select specific columns

```python
%%sql
SELECT FirstName, LastName, Department FROM Employees;
```

##### Select distinct values

Returns only unique values in the specified column(s).

```python
%%sql
SELECT DISTINCT Department FROM Employees;
```

#### `WHERE` Clause

Filters records based on specified conditions.

##### Basic filtering (equality, comparison operators)

```python
%%sql
SELECT * FROM Employees
WHERE Department = 'IT';

SELECT FirstName, Salary FROM Employees
WHERE Salary > 70000;
```

##### Using `AND`, `OR`, `NOT` operators

```python
%%sql
SELECT * FROM Employees
WHERE Department = 'HR' AND Salary > 50000;

SELECT * FROM Employees
WHERE Department = 'IT' OR Department = 'Marketing';

SELECT * FROM Employees
WHERE NOT Department = 'HR';
```

##### Using `LIKE` (pattern matching)

```python
%%sql
SELECT * FROM Employees
WHERE FirstName LIKE 'A%'; -- Starts with 'A'

SELECT * FROM Employees
WHERE LastName LIKE '%son%'; -- Contains 'son'

SELECT * FROM Employees
WHERE FirstName LIKE '_i%'; -- Second letter is 'i'
```

##### Using `IN`, `BETWEEN`, `IS NULL`

```python
%%sql
SELECT * FROM Employees
WHERE Department IN ('HR', 'IT');

SELECT * FROM Employees
WHERE Salary BETWEEN 60000 AND 70000;

SELECT * FROM Employees
WHERE Email IS NULL;
```

#### `ORDER BY`

Sorts the result set of a query based on one or more columns.

```python
%%sql
SELECT * FROM Employees
ORDER BY LastName ASC; -- Ascending order (default)

SELECT * FROM Employees
ORDER BY Salary DESC, FirstName ASC; -- Descending salary, then ascending first name
```

#### `LIMIT` (or `TOP` for SQL Server/MS Access)

Restricts the number of rows returned by the query.

```python
%%sql
SELECT * FROM Employees
LIMIT 3;
```

#### Aggregate Functions with `GROUP BY`

Aggregate functions (`COUNT`, `SUM`, `AVG`, `MIN`, `MAX`) perform calculations on a set of rows and return a single summary row. `GROUP BY` groups rows that have the same values in specified columns into summary rows.

```python
%%sql
SELECT Department, COUNT(EmployeeID) AS NumberOfEmployees
FROM Employees
GROUP BY Department;

SELECT Department, AVG(Salary) AS AverageSalary
FROM Employees
GROUP BY Department;
```

#### `HAVING` Clause

Filters groups based on specified conditions, used after `GROUP BY`.

```python
%%sql
SELECT Department, COUNT(EmployeeID) AS NumberOfEmployees
FROM Employees
GROUP BY Department
HAVING COUNT(EmployeeID) > 1;
```

#### `UPDATE`

Modifies existing records in a table.

```python
%%sql
UPDATE Employees
SET Salary = 85000.00
WHERE EmployeeID = 2;

UPDATE Employees
SET Department = 'IT Support'
WHERE Department = 'IT';
```

#### `DELETE FROM`

Deletes existing records from a table.

```python
%%sql
DELETE FROM Employees
WHERE EmployeeID = 1;

-- DELETE FROM Employees; -- Deletes all rows from the table
```

### 3. SQL Joins

Joins are used to combine rows from two or more tables, based on a related column between them. Let's create another table for demonstration.

```python
%%sql
CREATE TABLE Projects (
    ProjectID INT PRIMARY KEY,
    ProjectName VARCHAR(100) NOT NULL,
    EmployeeID INT,
    FOREIGN KEY (EmployeeID) REFERENCES Employees(EmployeeID)
);

INSERT INTO Projects (ProjectID, ProjectName, EmployeeID)
VALUES
    (101, 'Website Redesign', 2),
    (102, 'Mobile App Development', 5),
    (103, 'Marketing Campaign', 4),
    (104, 'HR System Upgrade', NULL); -- This project has no assigned employee yet
```

#### `INNER JOIN`

Returns rows when there is a match in both tables.

```python
%%sql
SELECT E.FirstName, E.LastName, P.ProjectName
FROM Employees E
INNER JOIN Projects P ON E.EmployeeID = P.EmployeeID;
```

#### `LEFT JOIN` (or `LEFT OUTER JOIN`)

Returns all rows from the left table, and the matching rows from the right table. If there is no match, the right side will have `NULL`s.

```python
%%sql
SELECT E.FirstName, E.LastName, P.ProjectName
FROM Employees E
LEFT JOIN Projects P ON E.EmployeeID = P.EmployeeID;
```

#### `RIGHT JOIN` (or `RIGHT OUTER JOIN`)

Returns all rows from the right table, and the matching rows from the left table. If there is no match, the left side will have `NULL`s.

```python
%%sql
SELECT E.FirstName, E.LastName, P.ProjectName
FROM Employees E
RIGHT JOIN Projects P ON E.EmployeeID = P.EmployeeID;
```

#### `FULL OUTER JOIN`

Returns all rows when there is a match in one of the tables. If there is no match, `NULL`s are filled in for the missing side.

```python
%%sql
-- Note: Not all SQL databases support FULL OUTER JOIN.
-- This query might need adjustment based on your specific database system (e.g., MySQL).
SELECT E.FirstName, E.LastName, P.ProjectName
FROM Employees E
FULL OUTER JOIN Projects P ON E.EmployeeID = P.EmployeeID;
```
